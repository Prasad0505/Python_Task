```python
import numpy as np
import pandas as pd

from datetime import datetime

import matplotlib.pyplot as plt
%matplotlib inline
```

Problem 1

Plot the total profit across the months

X label name = Month Number

Y label name = Total profit

```python
df=pd.read_csv('/content/drive/MyDrive/CSVFiles/company_data.csv')
df.head()
```

```python
plt.rcParms['figure.figsize']=(10,5)
```

```python
dfc=df[['total_profit']]
dfc

plt.plot(dfc)
plt.title('Total profit Across the month in Sales Company')
plt.ylabel('total_profit')
plt.xlabel('Month Number')
```

Problem 2

Read face cream and facewash product sales data and show it using the bar chart. 

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
face_df=df[['facecream','facewash']]

face_df.plot(kind='bar')
plt.title('sales per month')
plt.ylabel('Products')
plt.xlabel('Month')
```

Problem 3

Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```python
plt.plot(face_df[['facecream']],label='Face cream Sales Data',linewidth=3)
plt.plot(face_df[['facewash']],label='Face wash Sales Data',linewidth=3)
plt.plot(df[['toothpaste']],label='toothpaste Sales Data',linewidth=3)
plt.plot(df[['bathingsoap']],label='bathingsoap  Sales Data',linewidth=3)
plt.plot(df[['shampoo']],label='shampoo Sales Data',linewidth=3)
plt.plot(df[['moisturizer']],label='moisturizer Sales Data',linewidth=3)

plt.ylabel('Product')
plt.xlabel('Month')
plt.legend(loc='upper left')
```
